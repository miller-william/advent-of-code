```python
import pandas as pd
#read in and clean data
data = pd.read_csv('day-3-input.csv', header=None)
data = data.rename(columns={0: "full"})
#removing the erroneous last '/' character - a feature of way I copied the data across
data['full'] = data['full'].str[:12]
data
```

## Part 1

```python
gamma = ""
epsilon = ""

for i in range(0,12):
    #create new columns for each position and convert to numeric
    data[i] = pd.to_numeric(data['full'].str[i])
    #Test for which is most common value in column
    if data[i].sum()/len(data) >= 0.5:
        gamma += "1"
        epsilon += "0"
    else: 
        gamma += "0"
        epsilon += "1"
```

```python
data
```

```python
answer = int(gamma,2) * int(epsilon,2)
answer
```

## Part 2

```python
data2 = data
#Oxygen gen rating
for i in range(0,12):
    if data2[i].sum()/len(data2) >=0.5:
        #filter to only rows with most common value
        data2 = data2[data2[(i)] == 1]
    else:
        data2 = data2[data2[(i)] == 0]
        #stop process when we only have one record
    if len(data2) == 1:
        break

oxygen_gen_rating = int(data2.iloc[0,0],2)
oxygen_gen_rating
```

```python
data3 = data
#CO2 rating - same loop as above but take least common value in each column
for i in range(0,12):
    if data3[i].sum()/len(data3) >= 0.5:
        data3 = data3[data3[(i)] == 0]
    else:
        data3 = data3[data3[(i)] == 1]
    if len(data3) == 1:
        break

co2_rating = int(data3.iloc[0,0],2)
co2_rating
```

```python
answer2 = oxygen_gen_rating * co2_rating
answer2
```
