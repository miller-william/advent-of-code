```python
test_input = """11-22,95-115,998-1012,1188511880-1188511890,222220-222224,1698522-1698528,446443-446449,38593856-38593862,565653-565659,824824821-824824827,2121212118-2121212124"""
```

```python
def check_invalid(integer):
    if type(integer) != str:
        integer = str(integer)
    
    l_id = len(integer)

    if(l_id % 2 >0): # odd length
        return False
    
    l = int(l_id/2)

    if integer[:l] == integer[l:]:
        return True
```

```python
def p1(input):
    ans = 0
    input = [(int(a),int(b)) for a,b in [ full_id.split('-') for full_id in input.split(',')]] 
    for start,end in input:
        for id in range(start,end+1):
            if check_invalid(id):
                ans = ans + id
        
    return ans
```

```python
def check_invalid_seq(integer):
    if type(integer) != str:
        integer = str(integer)

    l_id = len(integer)
    l = int(l_id/2)

    for seq_len in range(l, 0, -1) :
        if l % seq_len > 0:
            next

        seq_valid = False # set to true as soon as we find a non symmetry 

        pattern = integer[0: seq_len]

        left = 0
        right = left + seq_len

        while right < l_id + 1 and not seq_valid: # iterate through the string comparisons
            left = left + seq_len
            right = right + seq_len

            if integer[left:right] != pattern:
                seq_valid = True # end this sequence of string comparisons because there's a mismatch
            if integer[left:right] == pattern and right == l_id: # we're at the end and it's all matched
                return True
    
    return False
```

```python
def p2(input):
    ans = 0
    input = [(int(a),int(b)) for a,b in [ full_id.split('-') for full_id in input.split(',')]] 
    for start,end in input:
        for id in range(start,end+1):
            if check_invalid_seq(id):
                ans = ans + id
        
    return ans
```
