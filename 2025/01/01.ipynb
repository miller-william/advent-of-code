```python
test_input = """L68
L30
R48
L5
R60
L55
L1
L99
R14
L82"""
test_input = test_input.splitlines()
```

```python
def p1(input):

    dir = {'L':-1,'R':1}

    instr = [ [dir.get(x[0]), int(x[1:])] for x in input]
    
    pos = 50
    ans = 0

    for dir, dist in instr:
        dist = dist % 100
        pos = pos + (dir * dist)
        if pos < 0:
            pos = 100 + pos
        if pos > 99:
            pos = pos % 100
        if pos in [0,100]: # landed on 0
            pos = 0
            ans += 1        

    return ans
```

```python
p1(test_input)
```

```python
p1(open('data01').read().splitlines())
```

```python
def p2(input):

    dir = {'L':-1,'R':1}

    instr = [ [dir.get(x[0]), int(x[1:])] for x in input]

    pos = 50
    ans = 0

    for dir, dist in instr:
        ans += dist // 100 # number of times the rotation loops fully round to the current position
        dist = dist % 100 # what's left after full loops

        new_pos = pos + (dir * dist) # new proposed position 
        if new_pos < 0: # if new position has crossed zero
            if pos != 0:
                ans += 1 # crossed 0
            pos = 100 + new_pos 
        elif new_pos > 100:
            ans += 1 # crossed 0
            pos = new_pos % 100
        elif new_pos in [0,100]: # landed on 0
            pos = 0
            ans += 1
        else:
            pos = new_pos        
            
    return ans
```

```python
p2(test_input)
```

```python
p2(open('data01').read().splitlines())
```
