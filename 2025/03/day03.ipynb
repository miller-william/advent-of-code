```python
test_input = """987654321111111
811111111111119
234234234234278
818181911112111"""
```

```python
def p1(input):
    ans = 0
    for bank in input.splitlines():
        l = len(bank)
        left = 0
        pos = None
        for i in range(l-1): #check all but the last integer
            integer_to_check = int(bank[i])
            if integer_to_check > left:
                left = integer_to_check
                pos = i
        
        right = 0
        for j in range(pos+1,l): # check integers to right
            integer_to_check = int(bank[j])
            if integer_to_check > right:
                right = integer_to_check
        
        ans = ans + int(str(left)+str(right))
    
    return ans
```

```python
p1(test_input)
```

```python
p1(open('data03').read())
```

```python
def p2(input, n):
    ans = 0
    for bank in input.splitlines():
        l = len(bank)
        joltage = n * ['']
        start_pos = 0

        for digit_position in range(0,n): 

            digit = 0
            max_position = l - ((n-1)-digit_position)

            for i in range(start_pos, max_position): 
                integer_to_check = int(bank[i])
                if integer_to_check > digit:
                    digit = integer_to_check
                    start_pos = i + 1
            joltage[digit_position] = str(digit)
        
        ans += int(''.join(joltage))

    return ans
```

```python
p2(test_input,12)
```

```python
p2(open('data03').read(), 12)
```
