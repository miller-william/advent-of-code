```python
test_input = """123 328  51 64 
 45 64  387 23 
  6 98  215 314
*   +   *   +  """
```

```python
def p1(input):
    main_ans = 0
    instr = [x.split() for x in input.splitlines()]

    for col in range(len(instr[0])): # each column
        operator = instr[-1][col]
        if operator == '*':
            ans = 1
            for row in range(len(instr)-1) :
                ans = ans * int(instr[row][col])
        else:
            ans = 0
            for row in range(len(instr)-1) :
                ans = ans + int(instr[row][col])

        main_ans = main_ans + ans

    return main_ans
```

```python
p1(test_input)
```

```python
p1(open('data06').read())
```

```python
def p2(input):

    nums = [[int(y) if y != ' ' else None for y in a ] for a in [list(x) for x in input.splitlines()[:-1]]]
    ops = input.splitlines()[-1].split()

    final_ans = 0

    op_pos = -1

    group = [] # the group of numbers for each operator

    for col in range(len(nums[0])-1,-1,-1):

        str_num = ''
        if all(row[col] is None for row in nums): # blank row - move to next number

            if ops[op_pos] == '*':
                ans = 1
                for n in group:
                    ans = ans * n
            else:
                ans = 0
                for n in group:
                    ans = ans + n
            
            final_ans = final_ans + ans
            
            op_pos = op_pos - 1 # move to next operator
            group = [] # start new group of nums
            continue

        for row in range(len(nums)):
            if not nums[row][col]: #current item is missing
                continue

            str_num = str_num + str(nums[row][col]) # build number for each group
        
        group.append(int(str_num)) # add complete number to the group

    if ops[op_pos] == '*':
        ans = 1
        for n in group:
            ans = ans * n
    else:
        ans = 0
        for n in group:
            ans = ans + n

    final_ans = final_ans + ans

    return final_ans
```

```python
p2(test_input)
```
