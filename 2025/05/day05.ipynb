```python
test_input = """3-5
10-14
16-20
12-18

1
5
8
11
17
32"""
```

```python
def p1(input):

    ranges, ing = [x.splitlines() for  x in input.split('\n\n')]
    ranges = [(int(a),int(b)) for a,b in [r.split('-') for r in ranges]]
    ing = [int(x) for x in ing]

    ans = 0
    for i in ing:
        for r in ranges:
            if r[0] <= i <= r[1]:
                ans +=1
                break
    return ans
```

```python
p1(test_input)
```

```python
def p2(input):

    ranges, ing = [x.splitlines() for  x in input.split('\n\n')]
    ranges = [(int(a),int(b)) for a,b in [r.split('-') for r in ranges]]

    ranges = sorted(ranges, key=lambda a: a[0])   # sort

    ans = 0
    for i in range(len(ranges)-1):
        l,r = ranges[i]
        n_l, n_r = ranges[i+1]

        if r > n_l and l < n_r:
            ranges[i+1] = (min(l, n_l), max(r, n_r))
        else:
            ans = ans + (r-l+1) # add distinct range
            
    l,r = ranges[i+1] # add final range
    ans = ans + (r-l+1)

    return ans
```

```python
p2(test_input)
```

```python
p2(open('data05').read())
```
