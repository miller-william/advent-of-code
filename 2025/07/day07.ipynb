```python
test_input = """.......S.......
...............
.......^.......
...............
......^.^......
...............
.....^.^.^.....
...............
....^.^...^....
...............
...^.^...^.^...
...............
..^...^.....^..
...............
.^.^.^.^.^...^.
..............."""
```

```python
def in_grid(pos, grid):
    n_rows = len(grid)
    n_cols = len(grid[0])

    if 0 <= pos[0] < n_rows and 0 <= pos[1] < n_cols:
        return True
    
    return False
```

```python
def beam(grid, start, visited):
    splits = set()

    current = grid[start[0]][start[1]]

    if start in visited:
        return set()

    visited.update({start})

    if current == '^': # Split happens

        splits.update({start})
        
        left = (start[0],start[1]-1)
        right = (start[0],start[1]+1)

        if in_grid(left, grid):
            splits.update(beam(grid, left, visited)) # start splits from left

        if in_grid(right, grid):
            if in_grid((right[0], right[1]+1),grid):
                if grid[right[0]][right[1]+1] != '^':
                    splits.update(beam(grid, right, visited)) # start splits from right
            else:
                splits.update(beam(grid, right, visited))

    if current in ['.','S']:
        below = (start[0]+1,start[1])

        if in_grid(below, grid):
            splits.update(beam(grid, below, visited)) # start splits from below
        
        else:
            return splits
    
    return splits
```

```python
def p1(input):

    grid = input.splitlines()
    S = (0,[i for i in range(len(grid[0])) if grid[0][i] == 'S'][0]) # find start

    return len(beam(grid, S, set()))
```

```python
from collections import defaultdict

def p2(input):
    grid = input.splitlines()
    beams = defaultdict(int)
    S = [i for i in range(len(grid[0])) if grid[0][i] == 'S'][0] # find start
    beams[S] = 1
    for row in grid:
        for i, s in enumerate(row):
            if s == '^' and beams.get(i,0) > 0:
                beams[i+1] += beams.get(i)
                beams[i-1] += beams.get(i)
                del beams[i]

    return (sum(beams.values()))
```
