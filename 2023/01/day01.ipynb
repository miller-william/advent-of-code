```python
import re
```

```python
example = """1abc2
pqr3stu8vwx
a1b2c3d4e5f
treb77uche9t"""
```

```python
def s(inp):
    ans = 0
    for row in inp.splitlines():
        num_list = re.findall("[0-9]",row)
        if len(num_list) < 2:
            num_list = num_list + num_list
        if len(num_list) > 2:
            num_list = num_list[0] + num_list[-1]
        ans = ans + int("".join(num_list))
    return(ans)
```

```python
s(example)
```

```python
##Â Part 2
ex2 = """two1nine
eightwothree
abcone2threexyz
xtwone3four
4nineeightseven2
zoneight234
7pqrstsixteen"""
```

```python
digits = {'one':'1',
         'two':'2',
         'three':'3',
         'four':'4',
         'five':'5',
         'six':'6',
         'seven':'7',
         'eight':'8',
         'nine':'9'}
```

```python
#replace first 'spelled digit' with number

def replace_first(s):
    first = len(s) + 1 
    # find the first 'spelled digit' in the string
    for k in digits:
        loc = s.find(k)
        if loc != -1 and loc < first: # if digit is present and before current
            first = loc
            num = k 
    if first < len(s) + 1:
        ln = len(num) - 1 # keep last letter in case it's used in last digit (awkward)
        s = s[:first] + digits[num] + s[first+ln:]
        return s
    else:
        return s
```

```python
replace_first("oonetwoone")
# see we retain the 'e' from 'one'
```

```python
#replace last
def replace_last(s):
    last = -2
    for k in digits:
        loc = s.rfind(k)
        if loc != -1 and loc > last:
            last = loc
            num = k 
    if last > -2:
        ln = len(num)
        s = s[:last] + digits[num] + s[last+ln:]
        return s
    else:
        return s
```

```python
replace_last("twoonetwo2one1one")
```

```python
# showing why we retain the last letter...
replace_last(replace_first("twone"))
```

```python
# add these two functions in to p1 solution
def s2(inp):
    ans = 0
    for row in inp.splitlines():
        print(row)
        row = replace_first(row)
        row = replace_last(row)
        num_list = re.findall("[0-9]",row)
        if len(num_list) < 2:
            num_list = num_list + num_list
        if len(num_list) > 2:
            num_list = num_list[0] + num_list[-1]
        print(int("".join(num_list)))
        ans = ans + int("".join(num_list))
    return(ans)
```
