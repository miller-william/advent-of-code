```python
instr = {
    "r": {"/" : "u", "O" : "d", "-" : "r", "|" : ["u","d"], "." : "r"},
    "l": {"/" : "d", "O" : "u", "-" : "l", "|" : ["u","d"], "." : "l"},
    "u": {"/" : "r", "O" : "l", "-" : ["l","r"], "|" : "u", "." : "u"},
    "d": {"/" : "l", "O" : "r", "-" : ["l","r"], "|" : "d", "." : "d"},
}
```

```python
def beam(idx, idy, grid, results, dir, splitter):
    # process direction
    while idx > -1 and idy > -1 and idx < len(grid[0]) and idy < len(grid): #while on grid
        char = (grid[idy][idx])
        new_dir = instr[dir][char]

        if len(new_dir) > 1: # beam has split
            if (idx,idy) not in results: # new splitter
                splitter.append((idx,idy))
                results = results.union(beam(idx,idy,grid,results,new_dir[0], splitter))
                results = results.union(beam(idx,idy,grid,results,new_dir[1], splitter))
                
            else:
                return results # end beam path

        else: # continue on with normal, non-splitting path
            results.add((idx,idy)) # add current position to the results
            dir = new_dir
            # move to next position
            if dir == "u":
                idy -= 1
            elif dir == "d":
                idy += 1
            elif dir == "l":
                idx -= 1
            elif dir == "r":
                idx += 1

        if idx < 0 or idy < 0 or idx == len(grid[0]) or idy == len(grid): #beam is off the edge of the grid:
            return results
```

```python
def p1(inp):
    results = set()
    inp = inp.replace('\\','O')
    grid = inp.split('\n')
    energised = beam(0,0,grid,results,'r', splitter=[])
    return(len(energised))
```

```python
def p1_for_p2(start_x,start_y,start_d, grid):
    results = set()
    energised = beam(start_x,start_y,grid,results,start_d, splitter=[])
    return(len(energised))
```

```python
def p2(inp):

    inp = inp.replace('\\','O')
    grid = inp.split('\n')
    
    best_ans = 0
    best_start = None

    #top row
    start_y = 0
    start_d = 'd'
    for start_x in range(0,len(grid[0])):
        new_ans = p1_for_p2(start_x,start_y,start_d, grid)
        if new_ans > best_ans:
            best_start = (start_x,start_y,start_d)
            best_ans = new_ans
            
    # bottom row
    start_y = len(grid) - 1 
    start_d = 'u'
    for start_x in range(0,len(grid[0])):
        new_ans = p1_for_p2(start_x,start_y,start_d, grid)
        if new_ans > best_ans:
            best_start = (start_x,start_y,start_d)
            best_ans = new_ans

    #left row - right
    start_x = 0
    start_d = 'r'
    for start_y in range(0,len(grid)):
        new_ans = p1_for_p2(start_x,start_y,start_d, grid)
        if new_ans > best_ans:
            best_ans = new_ans   

    #right row left
    start_x = len(grid[0]) - 1
    start_d = 'l'
    for start_y in range(0,len(grid)):
        new_ans = p1_for_p2(start_x,start_y,start_d, grid)
        if new_ans > best_ans:
            best_ans = new_ans
    
    return best_ans
    
```
